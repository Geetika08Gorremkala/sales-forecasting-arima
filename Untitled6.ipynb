```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and inspect data
df = pd.read_csv("mock_kaggle.csv")
df['data'] = pd.to_datetime(df['data'])  # convert to datetime
df = df.sort_values('data')

# Set date as index
df.set_index('data', inplace=True)

# Display summary
print(df.info())
df.head()
```

```python
# Resample to monthly sales
monthly_sales = df['venda'].resample('M').sum()

# Plot
plt.figure(figsize=(10, 4))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title("Monthly Sales Over Time")
plt.ylabel("Sales")
plt.xlabel("Month")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import numpy as np

# Split data into train/test
train = monthly_sales[:-3]
test = monthly_sales[-3:]

# Build and fit ARIMA model (auto tuning optional)
model = ARIMA(train, order=(1, 1, 1))
model_fit = model.fit()

# Forecast
forecast = model_fit.forecast(steps=3)
forecast.index = test.index

# Plot forecast vs actual
plt.figure(figsize=(10, 4))
plt.plot(train.index, train, label="Train")
plt.plot(test.index, test, label="Actual")
plt.plot(forecast.index, forecast, label="Forecast", linestyle='--')
plt.title("Sales Forecast")
plt.legend()
plt.grid(True)
plt.show()

# Evaluate
rmse = np.sqrt(mean_squared_error(test, forecast))
print(f"RMSE: {rmse:.2f}")
```

```python
# Save the monthly sales and forecast
results_df = pd.DataFrame({
    'actual': test,
    'forecast': forecast
})
results_df.to_csv("sales_forecast_results.csv")
# Optional: Save figure
plt.figure(figsize=(10, 4))
plt.plot(train.index, train, label="Train")
plt.plot(test.index, test, label="Actual")
plt.plot(forecast.index, forecast, label="Forecast", linestyle='--')
plt.title("Sales Forecast")
plt.legend()
plt.grid(True)
plt.savefig("forecast_plot.png")
plt.close()
```
